```python
#import dependents and packages to run notebook, import the csv file for data
import pandas as pd
import numpy as np
from IPython.display import display
file = 'Golf_Data.csv'
all_data_df = pd.read_csv(file)
year_var = input("Enter Desired Year: ")
main_data = all_data_df.loc[(all_data_df.Year.isin([year_var]))]
class font:
    BOLD = '\033[1m'
    END = '\033[0m'
```

#### Season Statistics

```python
#function for showing courses played and number of rounds played
def Courses_Played():
    courses = main_data['Course'].value_counts().reset_index()
    courses.rename(columns = {"index" : "Courses", "Course":"X"}, inplace = True)
    courses.drop(columns={'X'}, inplace=True)
    courses.sort_values(by='Courses', ascending = True, inplace=True)
    courses.reset_index(drop=True, inplace=True)
    print("         ")
    print('------------------------------------------------')
    print(f"{font.BOLD}Courses Played in {year_var}{font.END}")
    print('------------------------------------------------')
    display(courses)
    rd_count = main_data["Date"].nunique()
    print(f"Number of Rounds Played: {rd_count}")
    return
#Courses_Played()
```

```python
#function to display all 2022 Stats - print statements at end to show all
def All_Stats():
    total_holes = main_data['Hole'].count()
    total_score = main_data['Score'].sum()
    per_18 = (total_holes/18)
    avg_score_per18 = (total_score/per_18).round(2)
    total_par = main_data['Par'].sum()
    score_par_diff = total_score - total_par
    stroke_diff_per18 = (score_par_diff/per_18).round(2)
    score_sum = (main_data['Score']).sum()
    par_sum = (main_data['Par']).sum()
    hole_count = (main_data['Par']).count()
    score_to_par = ((score_sum - par_sum)/hole_count).round(2)
    tot_drives = main_data['FIR'].count()
    FIR = main_data['FIR'].sum()
    FIR_p = ((FIR/tot_drives)*100).round(2)
    tot_greens = main_data['GIR'].count()
    GIR = main_data['GIR'].sum()
    GIR_p = ((GIR/tot_greens)*100).round(2)
    putts_per_round = ((main_data['Putts'].sum())/per_18).round(1)
    times_putting = main_data['Putts'].count()
    one_putts_df = main_data.loc[main_data['Putts']==1]
    one_putts = one_putts_df["Putts"].count()
    one_putt_p = ((one_putts/times_putting)*100).round(2)
    two_putts_df = main_data.loc[main_data['Putts']==2]
    two_putts = two_putts_df["Putts"].count()
    two_putt_p = ((two_putts/times_putting)*100).round(2)
    three_putts_df = main_data.loc[main_data['Putts']>2]
    three_putts = three_putts_df["Putts"].count()
    three_putt_p = ((three_putts/times_putting)*100).round(2)
    tot_scramble = main_data['Scramble'].count()
    scramble_yes = main_data['Scramble'].sum()
    scramble_p = ((scramble_yes/tot_scramble)*100).round(2)
    par3_df = main_data.loc[main_data['Par']==3]
    par3_total_score = par3_df['Score'].sum()
    par3_holes_played = par3_df['Par'].count()
    par3_avg_score = (par3_total_score/par3_holes_played).round(2)
    par4_df = main_data.loc[main_data['Par']==4]
    par4_total_score = par4_df['Score'].sum()
    par4_holes_played = par4_df['Par'].count()
    par4_avg_score = (par4_total_score/par4_holes_played).round(2)
    par5_df = main_data.loc[main_data['Par']==5]
    par5_total_score = par5_df['Score'].sum()
    par5_holes_played = par5_df['Par'].count()
    par5_avg_score = (par5_total_score/par5_holes_played).round(2)
    names_df = main_data.drop(columns = {'Date', 'Month', 'Hole', 'FIR', 'GIR', 'Putts', 'Scramble'})
    names_df['Diff'] = (names_df['Score']-names_df['Par']).round(0)
    eagles = names_df.loc[names_df['Diff']==(-2.0)]
    num_eagles = eagles['Diff'].count()
    eag_perc = ((num_eagles/total_holes)*100).round(2)
    eagles_per18 = (num_eagles/per_18).round(2)
    birdies = names_df.loc[names_df['Diff']==(-1.0)]
    num_birdies = birdies['Diff'].count()
    birdie_perc = ((num_birdies/total_holes)*100).round(2)
    birdie_per18 = (num_birdies/per_18).round(2)
    pars = names_df.loc[names_df['Diff']==(0.0)]
    num_pars = pars['Diff'].count()
    par_perc = ((num_pars/total_holes)*100).round(2)
    par_per18 = (num_pars/per_18).round(2)
    bogeys = names_df.loc[names_df['Diff']==(1.0)]
    num_bogeys = bogeys['Diff'].count()
    bogey_perc = ((num_bogeys/total_holes)*100).round(2)
    bogey_per18 = (num_bogeys/per_18).round(2)
    dbl_bogeys = names_df.loc[names_df['Diff']==(2.0)]
    num_dblbogeys = dbl_bogeys['Diff'].count()
    dblboge_perc = ((num_dblbogeys/total_holes)*100).round(2)
    dblboge_per18 = (num_dblbogeys/per_18).round(2)
    trip_plus = names_df.loc[names_df['Diff']>(2.0)]
    num_tripbogeys = trip_plus['Diff'].count()
    tripbog_perc = ((num_tripbogeys/total_holes)*100).round(2)
    tripbog_per18 = (num_tripbogeys/per_18).round(2)
    FIR_missed_df = main_data.loc[main_data['FIR']==0]
    tot_miss_FIR = FIR_missed_df.FIR.count()
    tot_miss_FIRp = ((tot_miss_FIR/tot_drives)*100).round(2)
    FIR_mR = FIR_missed_df.loc[main_data['FIR_Miss']=='R']
    tot_FIR_mR = FIR_mR.FIR_Miss.count()
    FIR_mR_perc = ((tot_FIR_mR/tot_miss_FIR)*100).round(2)
    FIR_mL = FIR_missed_df.loc[main_data['FIR_Miss']=='L']
    tot_FIR_mL = FIR_mL.FIR_Miss.count()
    FIR_mL_perc = ((tot_FIR_mL/tot_miss_FIR)*100).round(2)
    GIR_missed_df = main_data.loc[main_data['GIR']==0]
    tot_missed_GIR = GIR_missed_df.GIR.count()
    tot_missed_GIRp = ((tot_missed_GIR/tot_greens)*100).round(2) 
    GIR_missR = GIR_missed_df.loc[GIR_missed_df['GIR_Miss']=='R']
    tot_GIR_missR = GIR_missR.GIR_Miss.count()
    tot_GIR_missRp = ((tot_GIR_missR/tot_missed_GIR)*100).round()    
    GIR_missL = GIR_missed_df.loc[GIR_missed_df['GIR_Miss']=='L']
    tot_GIR_missL = GIR_missL.GIR_Miss.count()
    tot_GIR_missLp = ((tot_GIR_missL/tot_missed_GIR)*100).round()    
    GIR_missS = GIR_missed_df.loc[GIR_missed_df['GIR_Miss']=='S']
    tot_GIR_missS = GIR_missS.GIR_Miss.count()
    tot_GIR_missSp = ((tot_GIR_missS/tot_missed_GIR)*100).round()    
    GIR_missI = GIR_missed_df.loc[GIR_missed_df['GIR_Miss']=='Icarito']
    tot_GIR_missI = GIR_missI.GIR_Miss.count()
    tot_GIR_missIp = ((tot_GIR_missI/tot_missed_GIR)*100).round()    
    print("                                      ")
    print('------------------------------------------------')
    print(f"{font.BOLD}All Stats for {year_var} Season{font.END}")
    print('------------------------------------------------')
    print(f"Average Score Per 18: {avg_score_per18}")
    print(f"Average {stroke_diff_per18} Strokes Over Par")
    print(f"Average Score to Par: {score_to_par}")
    print("     ")
    print(f"FIR and GIR Stats")
    print('------------------------------------------------')
    print(f"FIR Percentage: {FIR_p}%")
    print(f"GIR Percentage: {GIR_p}%")
    print(f"Total Number of Missed FIR: {tot_miss_FIR}")
    print(f"Missed FIR Percentage: {tot_miss_FIRp}%")
    print(f"Total Number of FIR Missed Right: {tot_FIR_mR}")
    print(f"Missed FIR Right Percentage: {FIR_mR_perc}%")
    print(f"Total Number of FIR Missed Left: {tot_FIR_mL}")
    print(f"Missed FIR Left Percentage: {FIR_mL_perc}%")
    print(f"Total Number of Missed GIR: {tot_missed_GIR}")
    print(f"Total Missed GIR Percentage: {tot_missed_GIRp}%")
    print(f"Total GIR Missed Right: {tot_GIR_missR}")
    print(f"Missed GIR Right Percentage: {tot_GIR_missRp}%")
    print(f"Total GIR Missed Left: {tot_GIR_missL}")
    print(f"Missed GIR Left Percentage: {tot_GIR_missLp}%")
    print(f"Total GIR Missed Short: {tot_GIR_missS}")
    print(f"Missed GIR Short Percentage: {tot_GIR_missSp}%")
    print(f"Total GIR Missed Long: {tot_GIR_missI}")
    print(f"Missed GIR Long Percentage: {tot_GIR_missIp}%")
    print("             ")
    print("Putting and Scramble Stats")
    print('------------------------------------------------')
    print(f"Average Putts Per Round: {putts_per_round}")
    print(f"One Putt Frequency: {one_putt_p}%")
    print(f"Two Putt Frequency: {two_putt_p}%")
    print(f"Three Putt or Worse Frequency: {three_putt_p}%")
    print(f"Scramble Conversion: {scramble_p}%")
    print("          ")
    print("Scoring Stats")
    print('------------------------------------------------')
    print(f"Average Par 3 Scoring: {par3_avg_score}")
    print(f"Average Par 4 Scoring: {par4_avg_score}")
    print(f"Average Par 5 Scoring: {par5_avg_score}")
    print(f"Total Eagles: {num_eagles}")
    print(f"Eagle Percentage: {eag_perc}%")
    print(f"Eagles Per 18 Holes: {eagles_per18}")
    print(f"Total Birdies: {num_birdies}")
    print(f"Birdie Percentage: {birdie_perc}%")
    print(f"Birdies Per 18 Holes: {birdie_per18}")
    print(f"Total Pars: {num_pars}")
    print(f"Par Percentage: {par_perc}%")
    print(f"Pars Per 18 Holes: {par_per18}")
    print(f"Total Bogeys: {num_bogeys}")
    print(f"Bogey Percentage: {bogey_perc}%")
    print(f"Bogeys Per 18 Holes: {bogey_per18}")
    print(f"Total Double Bogeys: {num_dblbogeys}")
    print(f"Double Bogey Percentage: {dblboge_perc}%")
    print(f"Double Bogeys Per 18 Holes: {dblboge_per18}")
    print(f"Total Triple Bogeys or Worse: {num_tripbogeys}")
    print(f"Triple Bogey or Worse Percentage: {tripbog_perc}%")
    print(f"Triple Bogeys or Worse Per 18 Holes: {tripbog_per18}")
    return
#All_Stats()
```

```python
def Selected_Round():
    print("           ")
    print('------------------------------------------------')
    print(font.BOLD + 'SINGLE ROUND STATS' + font.END)
    print('------------------------------------------------')
    rnd = input("Enter Desired Round Date: ")
    rnd_df = main_data.loc[(main_data.Date.isin([rnd]))]
    rnd_course = rnd_df.iat[0,3]
    score_rnd = rnd_df['Score'].sum()
    rnd_tot_par = rnd_df.Par.sum()
    rnd_score_par_diff = score_rnd - rnd_tot_par
    rnd_hole_count = rnd_df.Par.count()
    rnd_score_topar = ((score_rnd-rnd_tot_par)/rnd_hole_count).round(2)
    putts_rnd = rnd_df['Putts'].sum()
    one_putts_df = rnd_df.loc[rnd_df['Putts']== 1]
    one_putts = one_putts_df['Putts'].count()
    two_putts_df = rnd_df.loc[rnd_df['Putts']== 2]
    two_putts = two_putts_df['Putts'].count()
    three_putts_df = rnd_df.loc[rnd_df['Putts']== 3]
    three_putts = three_putts_df['Putts'].count()
    fir_num = rnd_df['FIR'].count()
    fir_hit = rnd_df['FIR'].sum()
    fir_hitperc = ((fir_hit/fir_num)*100).round(2)
    gir_num = rnd_df['GIR'].count()
    gir_hit = rnd_df['GIR'].sum()
    gir_hitperc = ((gir_hit/gir_num)*100).round(2)
    rnd_FIR_missed_df = rnd_df.loc[rnd_df['FIR']==0]
    rnd_tot_miss_FIR = rnd_FIR_missed_df.FIR.count()
    rnd_tot_miss_FIRp = ((rnd_tot_miss_FIR/fir_num)*100).round(2)
    rnd_FIR_mR = rnd_FIR_missed_df.loc[rnd_FIR_missed_df['FIR_Miss']=='R']
    rnd_tot_FIR_mR = rnd_FIR_mR.FIR_Miss.count()
    rnd_FIR_mR_perc = ((rnd_tot_FIR_mR/rnd_tot_miss_FIR)*100).round(2)
    rnd_FIR_mL = rnd_FIR_missed_df.loc[rnd_FIR_missed_df['FIR_Miss']=='L']
    rnd_tot_FIR_mL = rnd_FIR_mL.FIR_Miss.count()
    rnd_FIR_mL_perc = ((rnd_tot_FIR_mL/rnd_tot_miss_FIR)*100).round(2)
    rnd_GIR_missed_df = rnd_df.loc[rnd_df['GIR']==0]
    rnd_tot_missed_GIR = rnd_GIR_missed_df.GIR.count()
    rnd_tot_missed_GIRp = ((rnd_tot_missed_GIR/gir_num)*100).round(2)
    rnd_GIR_missR = rnd_GIR_missed_df.loc[rnd_GIR_missed_df['GIR_Miss']=='R']
    rnd_tot_GIR_missR = rnd_GIR_missR.GIR_Miss.count()
    rnd_tot_GIR_missRp = ((rnd_tot_GIR_missR/rnd_tot_missed_GIR)*100).round()    
    rnd_GIR_missL = rnd_GIR_missed_df.loc[rnd_GIR_missed_df['GIR_Miss']=='L']
    rnd_tot_GIR_missL = rnd_GIR_missL.GIR_Miss.count()
    rnd_tot_GIR_missLp = ((rnd_tot_GIR_missL/rnd_tot_missed_GIR)*100).round()
    rnd_GIR_missS = rnd_GIR_missed_df.loc[rnd_GIR_missed_df['GIR_Miss']=='S']
    rnd_tot_GIR_missS = rnd_GIR_missS.GIR_Miss.count()
    rnd_tot_GIR_missSp = ((rnd_tot_GIR_missS/rnd_tot_missed_GIR)*100).round()    
    rnd_GIR_missI = rnd_GIR_missed_df.loc[rnd_GIR_missed_df['GIR_Miss']=='Icarito']
    rnd_tot_GIR_missI = rnd_GIR_missI.GIR_Miss.count()
    rnd_tot_GIR_missIp = ((rnd_tot_GIR_missI/rnd_tot_missed_GIR)*100).round()    
    scram_num = rnd_df['Scramble'].count()
    scram_y = rnd_df['Scramble'].sum()
    scram_conv = ((scram_y/scram_num)*100).round(2)
    rndNames_df = rnd_df.drop(columns = {'Date', 'Month', 'Hole', 'FIR', 'GIR', 'Putts', 'Scramble'})
    rndNames_df['Diff'] = (rndNames_df['Score']-rndNames_df['Par']).round(0)
    rnd_eag_df = rndNames_df.loc[rndNames_df['Diff']== (-2)]
    rnd_eag = rnd_eag_df['Diff'].count()
    rnd_bird_df = rndNames_df.loc[rndNames_df['Diff']== (-1)]
    rnd_bird = rnd_bird_df['Diff'].count()
    rnd_par_df = rndNames_df.loc[rndNames_df['Diff']== (0)]
    rnd_par = rnd_par_df['Diff'].count()
    rnd_boge_df = rndNames_df.loc[rndNames_df['Diff']== (1)]
    rnd_boge = rnd_boge_df['Diff'].count()
    rnd_dblb_df = rndNames_df.loc[rndNames_df['Diff']== (2)]
    rnd_dblb = rnd_dblb_df['Diff'].count()
    rnd_tripbw_df = rndNames_df.loc[rndNames_df['Diff'] > (2)]
    rnd_tripbw = rnd_tripbw_df['Diff'].count()
    rnd_eag_p = ((rnd_eag/rnd_hole_count)*100).round(2)
    rnd_bird_p = ((rnd_bird/rnd_hole_count)*100).round(2)
    rnd_par_p = ((rnd_par/rnd_hole_count)*100).round(2)
    rnd_boge_p = ((rnd_boge/rnd_hole_count)*100).round(2)
    rnd_dblb_p = ((rnd_dblb/rnd_hole_count)*100).round(2)
    rnd_tripbw_p = ((rnd_tripbw/rnd_hole_count)*100).round(2)
    print("          ")
    print('------------------------------------------------')
    print(f'{font.BOLD}Stats for Round on: {rnd} at {rnd_course} {font.END}')
    print('------------------------------------------------')
    print(f"Score: {score_rnd}")
    print(f"Strokes Over Par: {rnd_score_par_diff}")
    print(f"Score to Par Per Hole: {rnd_score_topar}")
    print("          ")
    print("FIR and GIR Stats")
    print('------------------------------------------------')
    print(f"Fairways Hit: {fir_hit}")
    print(f"Fairways Hit Percentage: {fir_hitperc}%")
    print(f"Greens Hit: {gir_hit}")
    print(f"Green Hit Percentage: {gir_hitperc}%")
    print(f"Total Number of Missed FIR: {rnd_tot_miss_FIR}")
    print(f"Missed FIR Percentage: {rnd_tot_miss_FIRp}%")
    print(f"Total Number of FIR Missed Right: {rnd_tot_FIR_mR}")
    print(f"Missed FIR Right Percentage: {rnd_FIR_mR_perc}%")
    print(f"Total Number of FIR Missed Left: {rnd_tot_FIR_mL}")
    print(f"Missed FIR Left Percentage: {rnd_FIR_mL_perc}%")
    print(f"Total Number of Missed GIR: {rnd_tot_missed_GIR}")
    print(f"Total Missed GIR Percentage: {rnd_tot_missed_GIRp}%")
    print(f"Total GIR Missed Right: {rnd_tot_GIR_missR}")
    print(f"Missed GIR Right Percentage: {rnd_tot_GIR_missRp}%")
    print(f"Total GIR Missed Left: {rnd_tot_GIR_missL}")
    print(f"Missed GIR Left Percentage: {rnd_tot_GIR_missLp}%")
    print(f"Total GIR Missed Short: {rnd_tot_GIR_missS}")
    print(f"Missed GIR Short Percentage: {rnd_tot_GIR_missSp}%")
    print(f"Total GIR Missed Long: {rnd_tot_GIR_missI}")
    print(f"Missed GIR Long Percentage: {rnd_tot_GIR_missIp}%")
    print('                  ')
    print("Putting and Scramble Stats")
    print('------------------------------------------------')
    print(f"Number of Putts: {putts_rnd}")
    print(f"Number of One Putts: {one_putts}")
    print(f"Number of Two Putts: {two_putts}")
    print(f"Number of Three Putts: {three_putts}")
    print(f"Scramble Conversion: {scram_conv}%")
    print("          ")
    print("Scoring Stats")
    print('------------------------------------------------')
    print(f"Eagles Made: {rnd_eag}")
    print(f"Birdies Made: {rnd_bird}")
    print(f"Pars Made: {rnd_par}")
    print(f"Bogeys Made: {rnd_boge}")
    print(f"Double Bogeys Made: {rnd_dblb}")
    print(f"Triple Bogeys or Worse Made: {rnd_tripbw}")
    print(f"Eagles Made Percentage: {rnd_eag_p}%")
    print(f"Birdies Made Percentage: {rnd_bird_p}%")
    print(f"Pars Made Percentage: {rnd_par_p}%")
    print(f"Bogeys Made Percentage: {rnd_boge_p}%")
    print(f"Double Bogeys Made Percentage: {rnd_dblb_p}%")
    print(f"Triple Bogeys or Worse Made Percentage: {rnd_tripbw_p}%")
    return
```

```python
def Selected_Months():
    print("               ")
    print('------------------------------------------------')
    print(font.BOLD + 'SELECTED MONTH STATS' + font.END)
    print('------------------------------------------------')
    month_1 = input("Enter Desired Month ")
    month_2 = input("Enter Desired Month ")
    month_3 = input("Enter Desired Month ")
    month_4 = input("Enter Desired Month ")
    months = main_data.loc[(main_data.Month.isin([month_1, month_2, month_3, month_4]))]
    mo_rd_count = months["Date"].nunique()
    month_holes_played = months.Hole.count()
    months_totscore = months.Score.sum()
    months_per18 = (month_holes_played/18)
    months_score_per18 = (months_totscore/months_per18).round(2)
    month_totalpar = months.Par.sum()
    par_score_diff = months_totscore - month_totalpar
    month_avdiff = (par_score_diff/months_per18).round(2)
    months_stp = ((months_totscore-month_totalpar)/month_holes_played).round(2)
    mo_drives = months.FIR.count()
    mo_FIR = months.FIR.sum()
    mo_FIRp = ((mo_FIR/mo_drives)*100).round(2)
    mo_greens = months.GIR.count()
    mo_GIR = months.GIR.sum()
    mo_GIRp = ((mo_GIR/mo_greens)*100).round(2)
    mo_FIR_missed_df = months.loc[months['FIR']==0]
    mo_tot_miss_FIR = mo_FIR_missed_df.FIR.count()
    mo_tot_miss_FIRp = ((mo_tot_miss_FIR/mo_drives)*100).round(2)
    mo_FIR_mR = mo_FIR_missed_df.loc[mo_FIR_missed_df['FIR_Miss']=='R']
    mo_tot_FIR_mR = mo_FIR_mR.FIR_Miss.count()
    mo_FIR_mR_perc = ((mo_tot_FIR_mR/mo_tot_miss_FIR)*100).round(2)    
    mo_FIR_mL = mo_FIR_missed_df.loc[mo_FIR_missed_df['FIR_Miss']=='L']
    mo_tot_FIR_mL = mo_FIR_mL.FIR_Miss.count()
    mo_FIR_mL_perc = ((mo_tot_FIR_mL/mo_tot_miss_FIR)*100).round(2)    
    mo_GIR_missed_df = months.loc[months['GIR']==0]
    mo_tot_missed_GIR = mo_GIR_missed_df.GIR.count()
    mo_tot_missed_GIRp = ((mo_tot_missed_GIR/mo_greens)*100).round(2)    
    mo_GIR_missR = mo_GIR_missed_df.loc[mo_GIR_missed_df['GIR_Miss']=='R']
    mo_tot_GIR_missR = mo_GIR_missR.GIR_Miss.count()
    mo_tot_GIR_missRp = ((mo_tot_GIR_missR/mo_tot_missed_GIR)*100).round()    
    mo_GIR_missL = mo_GIR_missed_df.loc[mo_GIR_missed_df['GIR_Miss']=='L']
    mo_tot_GIR_missL = mo_GIR_missL.GIR_Miss.count()
    mo_tot_GIR_missLp = ((mo_tot_GIR_missL/mo_tot_missed_GIR)*100).round()    
    mo_GIR_missS = mo_GIR_missed_df.loc[mo_GIR_missed_df['GIR_Miss']=='S']
    mo_tot_GIR_missS = mo_GIR_missS.GIR_Miss.count()
    mo_tot_GIR_missSp = ((mo_tot_GIR_missS/mo_tot_missed_GIR)*100).round()    
    mo_GIR_missI = mo_GIR_missed_df.loc[mo_GIR_missed_df['GIR_Miss']=='Icarito']
    mo_tot_GIR_missI = mo_GIR_missI.GIR_Miss.count()
    mo_tot_GIR_missIp = ((mo_tot_GIR_missI/mo_tot_missed_GIR)*100).round() 
    mo_putts_per18 = ((months.Putts.sum())/months_per18).round(1)
    mo_tot_putts = months.Putts.count()
    mo_onepdf = months.loc[months['Putts']==1]
    mo_1pt = mo_onepdf.Putts.count()
    mo_1pt_p = ((mo_1pt/mo_tot_putts)*100).round(2)
    mo_twopdf = months.loc[months['Putts']==2]
    mo_2pt = mo_twopdf.Putts.count()
    mo_2pt_p = ((mo_2pt/mo_tot_putts)*100).round(2)
    mo_threepdf = months.loc[months['Putts']==3]
    mo_3pt = mo_threepdf.Putts.count()
    mo_3pt_p = ((mo_3pt/mo_tot_putts)*100).round(2)
    mo_tot_scram = months.Scramble.count()
    mo_scramY = months.Scramble.sum()
    mo_scramp = ((mo_scramY/mo_tot_scram)*100).round(2)
    mo_p3_df = months.loc[months['Par']==3]
    mo_p3_totS = mo_p3_df.Score.sum()
    mo_p3_totP = mo_p3_df.Par.count()
    mo_p3_AvgS = (mo_p3_totS/mo_p3_totP).round(2)
    mo_p4_df = months.loc[months['Par']==4]
    mo_p4_totS = mo_p4_df.Score.sum()
    mo_p4_totP = mo_p4_df.Par.count()
    mo_p4_AvgS = (mo_p4_totS/mo_p4_totP).round(2)
    mo_p5_df = months.loc[months['Par']==5]
    mo_p5_totS = mo_p5_df.Score.sum()
    mo_p5_totP = mo_p5_df.Par.count()
    mo_p5_AvgS = (mo_p5_totS/mo_p5_totP).round(2)
    mo_names = months.drop(columns = {'Date', 'Month', 'Hole', 'FIR', 'GIR', 'Putts', 'Scramble'})
    mo_names['Diff'] = (mo_names['Score']-mo_names['Par'])
    mo_Eag = mo_names.loc[mo_names['Diff']==(-2)]
    mo_E = mo_Eag.Diff.count()
    mo_Ep = ((mo_E/month_holes_played)*100).round(2)
    mo_Ep18 = (mo_E/months_per18).round(2)
    mo_Bird = mo_names.loc[mo_names['Diff']==(-1)]
    mo_B = mo_Bird.Diff.count()
    mo_Bp = ((mo_B/month_holes_played)*100).round(2)
    mo_Bp18 = (mo_B/months_per18).round(2)
    mo_Par = mo_names.loc[mo_names['Diff']==(0)]
    mo_P = mo_Par.Diff.count()
    mo_Pp = ((mo_P/month_holes_played)*100).round(2)
    mo_Pp18 = (mo_P/months_per18).round(2)
    mo_Bog = mo_names.loc[mo_names['Diff']==(1)]
    mo_Bo = mo_Bog.Diff.count()
    mo_Bop = ((mo_Bo/month_holes_played)*100).round(2)
    mo_Bop18 = (mo_Bo/months_per18).round(2)
    mo_DBog = mo_names.loc[mo_names['Diff']==(2)]
    mo_DBo = mo_DBog.Diff.count()
    mo_DBop = ((mo_DBo/month_holes_played)*100).round(2)
    mo_DBop18 = (mo_DBo/months_per18).round(2)
    mo_TBog = mo_names.loc[mo_names['Diff']>(2)]
    mo_TBo = mo_TBog.Diff.count()
    mo_TBop = ((mo_TBo/month_holes_played)*100).round(2)
    mo_TBop18 = (mo_TBo/months_per18).round(2)
    print("                                      ")
    print('------------------------------------------------')
    if month_1 and month_2 and month_3 and month_4:
        print(f"{font.BOLD}Stats for {month_1}, {month_2}, {month_3}, and {month_4}{font.END}")
    elif month_1 and month_2 and month_3:
        print(f"{font.BOLD}Stats for {month_1}, {month_2}, and {month_3}{font.END}")
    elif month_1 and month_2:
        print(f"{font.BOLD}Stats for {month_1} and {month_2}{font.END}")
    elif month_1:
        print(f"{font.BOLD}Stats for {month_1}{font.END}")
    print('------------------------------------------------')
    print(f"Number of Rounds Played: {mo_rd_count}")
    print(f"Average Score Per 18: {months_score_per18}")
    print(f"Averaged {month_avdiff} Strokes Over Par Per 18")
    print(f"Average Score to Par: {months_stp}")
    print("                ")
    print("FIR and GIR Stats")
    print('------------------------------------------------')
    print(f"Number of FIR: {mo_FIR}")
    print(f"FIR Percentage: {mo_FIRp}%")
    print(f"Number of GIR: {mo_GIR}")
    print(f"GIR Percentage: {mo_GIRp}%")
    print(f"Total Number of Missed FIR: {mo_tot_miss_FIR}")
    print(f"Missed FIR Percentage: {mo_tot_miss_FIRp}%")
    print(f"Total Number of FIR Missed Right: {mo_tot_FIR_mR}")
    print(f"Missed FIR Right Percentage: {mo_FIR_mR_perc}%")
    print(f"Total Number of FIR Missed Left: {mo_tot_FIR_mL}")
    print(f"Missed FIR Left Percentage: {mo_FIR_mL_perc}%")
    print(f"Total Number of Missed GIR: {mo_tot_missed_GIR}")
    print(f"Total Missed GIR Percentage: {mo_tot_missed_GIRp}%")
    print(f"Total GIR Missed Right: {mo_tot_GIR_missR}")
    print(f"Missed GIR Right Percentage: {mo_tot_GIR_missRp}%")
    print(f"Total GIR Missed Left: {mo_tot_GIR_missL}")
    print(f"Missed GIR Left Percentage: {mo_tot_GIR_missLp}%")
    print(f"Total GIR Missed Short: {mo_tot_GIR_missS}")
    print(f"Missed GIR Short Percentage: {mo_tot_GIR_missSp}%")
    print(f"Total GIR Missed Long: {mo_tot_GIR_missI}")
    print(f"Missed GIR Long Percentage: {mo_tot_GIR_missIp}%")
    print("                ")
    print("Putting and Scramble Stats")
    print('------------------------------------------------')
    print(f"Average Putts Per Round: {mo_putts_per18}")
    print(f"One Putt Frequency: {mo_1pt_p}%")
    print(f"Two Putt Frequency: {mo_2pt_p}%")
    print(f"Three Putt Frequency: {mo_3pt_p}%")
    print(f"Scramble Conversion: {mo_scramp}%")
    print("                 ")
    print("Scoring Stats")
    print('------------------------------------------------')
    print(f"Average Par 3 Scoring: {mo_p3_AvgS}")
    print(f"Average Par 4 Scoring: {mo_p4_AvgS}")
    print(f"Average Par 5 Scoring: {mo_p5_AvgS}")
    print(f"Total Eagles: {mo_E}")
    print(f"Eagle Percentage: {mo_Ep}%")
    print(f"Eagles Per 18 Holes: {mo_Ep18}")
    print(f"Total Birdies: {mo_B}")
    print(f"Birdie Percentage: {mo_Bp}%")
    print(f"Birdies Per 18 Holes: {mo_Bp18}")
    print(f"Total Pars: {mo_P}")
    print(f"Par Percentage: {mo_Pp}%")
    print(f"Pars Per 18 Holes: {mo_Pp18}")
    print(f"Total Bogeys: {mo_Bo}")
    print(f"Bogey Percentage: {mo_Bop}%")
    print(f"Bogeys Per 18 Holes: {mo_Bop18}")
    print(f"Total Double Bogeys: {mo_DBo}")
    print(f"Double Bogey Percentage: {mo_DBop}%")
    print(f"Double Bogeys Per 18 Holes: {mo_DBop18}")
    print(f"Total Triple Bogeys or Worse: {mo_TBo}")
    print(f"Triple Bogey or Worse Percentage: {mo_TBop}%")
    print(f"Triple Bogeys or Worse Per 18 Holes: {mo_TBop18}")
    return   
```
